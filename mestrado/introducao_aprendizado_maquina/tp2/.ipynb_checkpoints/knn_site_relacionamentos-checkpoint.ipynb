### 4 - Normalizando os dados

Perceba que a quantidade de milhas a√©reas domina o algoritmo de aprendizado, caso n√£o haja normaliza√ß√£o

```python
data.head()
```

```python
def normalize(X_train):
    
    # Min-Max Scaling
    m_train = X_train.shape[0]
    min_val = np.min(X_train, axis=0)
    max_val = np.max(X_train, axis=0)
    range_val = max_val - min_val
    
    normalized_dataset = np.zeros(X_train.shape)
    normalized_dataset = X_train - np.tile(min_val, (m_train, 1))
    normalized_dataset = normalized_dataset / np.tile(range_val, (m_train, 1))
    
#     # Z-score Standardization
#     mu = np.zeros(X_train.shape[1])
#     std = np.zeros(X_train.shape[1])
    
#     mu = np.array([np.mean(x) for x in X_train.T]) #calculando a m√©dia das 3 features
#     std = np.array([np.std(x) for x in X_train.T]) #calculando o desvio padr√£o das 3 features
    
#     X_train = X_train - np.tile(mu, (m_train,1))# np.tile repete o vetor mu=[mu_milhas, mu_games, mu_sorvete] m vezes
#     X_train = X_train / np.tile(std, (m_train,1))
    
    # return X_train, mu, sigma
    return normalized_dataset, min_val, range_val

X_train, min_val, range_val = normalize(X_train)
min_val
```

### 6 - Classificando uma nova amostra

```python
def classify(new_x, X_train, y_train, k):
    
    m = X_train.shape[0]
    
    #Calcular as dist√¢ncias
    diff_absoluta = np.tile(new_x, (m,1)) - X_train
    distancias = np.sqrt(np.sum(diff_absoluta**2, axis=1))
    
    #Obter o rank das k menores dist√¢ncias
    k_indices = np.argsort(distancias)[:k]
    labels_k_vizinhos = y_train[k_indices]
    
    #Classificar pelo voto majorit√°rio
    classnames, id_classes = np.unique(labels_k_vizinhos, return_inverse=True)
    counts = np.zeros(len(classnames)) #Quantidade de diferentes labels dentre os k vizinhos
    for i, classe in enumerate(classnames):
        counts[i] = np.sum(labels_k_vizinhos == classe)
    indice_classe_majoritaria = np.argmax(counts)
    return classnames[indice_classe_majoritaria]
```

```python
novo_x = np.array([68255, 11.638205, 0.409125])
novo_x = (novo_x - min_val) / range_val
print(novo_x)
y_hat = classify(novo_x, X_train, y_train, k=3)
print('Classe predita: {}'.format(y_hat))
```

### 7 - Calculando a acur√°cia

```python
k = 20

y_hat = [classify(((new_x-min_val)/range_val), X_train, y_train, k) for new_x in X_test]
accuracy = np.sum(y_hat == y_test) / float(len(y_test))
print("Acur√°cia para os dados de teste= %.2f" % accuracy)
```
