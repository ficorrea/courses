### Importante pacotes necess√°rios

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.image import imread
#import cv2
 
%matplotlib inline
```

### Visualizando a imagem

```python
# Lendo imagem
image = imread('pequi.jpg')
plt.imshow(image)
np.shape(image)
```

### Adequando a imagem para aplica√ß√£o do K-means

```python
vec_img = image.reshape((-1, 3))
print(image[0:10, :])
vec_img = np.float32(vec_img)
```

### Definindo M√©trica de Dist√¢ncia

```python
def distEuclidean(vecA, vecB):
    return np.sqrt(np.sum((vecA-vecB)**2))
```

### Gerando os centr√≥ides iniciais

```python
def randCent(dataSet, k):
    m, n = np.shape(dataSet) #Buscando qtde dimens√µes de um centr√≥ide (n)
    centroids = np.zeros((k,n)) #Criando a matriz com centr√≥ides
    indices_m = np.random.randint(m, size=k) #sorteia k amostras aleatoriamente para serem centr√≥ides
    centroids = dataSet[indices_m]
    return centroids
```

## Algoritmo K-means

```python
def kMeans(dataSet, k, max_iter=10, distMeas=distEuclidean, createCent=randCent):
    m = np.shape(dataSet)[0]
    clusterAssment = np.zeros((m,2)) # Matriz para atribuir as amostras aos clusters(coluna 1), a coluna 2 armazena a dist√¢ncia da amostra ao centr√≥ide
    centroids = createCent(dataSet, k)
    clusterChanged = True
    it = 0
    while clusterChanged and it < max_iter: #Crit√©rio de parada -> nenhuma amostra mudar de cluster
        clusterChanged = False
        #Passo 1 -> Atribui√ß√£o de Clusters
        for i in range(m):# Para cada amostra atribuir o ind√≠ce do cluster com centr√≥ide mais pr√≥ximo
            minDist = np.inf;
            minIndex = -1
            for j in range(k):
                distJI = distMeas(centroids[j,:],dataSet[i,:])
                if distJI < minDist:
                    minDist = distJI
                    minIndex = j
            if clusterAssment[i,0] != minIndex:
                clusterChanged = True
            clusterAssment[i,:] = minIndex, minDist**2
        
        #Passo 2 -> Recalculando os centr√≥ides
        for cent in range(k):
            ptsInClust = dataSet[clusterAssment[:,0] == cent] #Obtem todos os dados de um cluster
            centroids[cent,:] = np.mean(ptsInClust, axis=0) #Muda o centr√≥ide para a m√©dia 
        it += 1
    return centroids, clusterAssment
```

### Executando o K-means

```python
k = 12
max_iter = 100
centroids, clusterAssment = kMeans(vec_img, k, max_iter)
C = np.int32(clusterAssment[:,0]) #Vetor contendo o cluster das amostras
```

### Gerando a nova imagem

```python
centroids = np.uint8(centroids) #tornando os centr√≥ides valores inteiros
new_img = centroids[C] #Criando a nova imagem com os centr√≥ides
new_img = new_img.reshape((image.shape))
plt.imshow(new_img)
```
