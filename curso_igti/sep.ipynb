```
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```
df = pd.read_csv('train.csv', header=None)
```

```
plt.figure(figsize=(12, 8))
sns.heatmap(df.corr())
plt.show()
```

```
hg = []
for i in range(0, 500):
    for x in range(0, 500):
        c = df.corr()[i][x]
        if c >= 0.85 or c <= -0.85:
            hg.append([i, x, c])
```
