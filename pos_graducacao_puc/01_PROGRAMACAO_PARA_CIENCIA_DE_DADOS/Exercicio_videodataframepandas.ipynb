```python
import pandas as pd
import numpy as np
```

```python
# Dataframe com 12 linhas e 6 colunas
data = np.reshape(np.arange(1, 73), (12,6))
columns = ['A', 'B', 'C', 'D', 'E', 'F']
df = pd.DataFrame(data, columns=columns)
df
```

```python
# SubTotal1 com soma A, C e E
df['SubTotal1'] = df.apply(lambda row: sum([row['A'], row['C'], row['E']]), axis=1)
df
```

```python
# SubTotal1 com soma B, D e F
df['SubTotal2'] = df.apply(lambda row: sum([row['B'], row['D'], row['F']]), axis=1)
df
```

```python
# Total com soma de todas as colunas
df['Total'] = df.apply(lambda row: sum([row['SubTotal1'], row['SubTotal2']]), axis=1)
df
```

```python
# DF2 com colunas SubTotal1 e SubTotal2
df2 = df.loc[:, ['SubTotal1', 'SubTotal2']]
df2
```

```python
# Excluir subs df
df.drop(columns=['SubTotal1', 'SubTotal2'], inplace=True)
df
```

```python
# DF3 com 4 Ãºltimas linhas e 3 primeiras colunas
df3 = df.iloc[-4:, 0:3]
df3
```

```python
# Soma das linhas no final
# --> df = df.append(df.sum, ignore_index=True)
somas = pd.Series([df[i].sum() for i in df.columns], index=df.columns)
df = df.append(somas, ignore_index=True)
df
```
